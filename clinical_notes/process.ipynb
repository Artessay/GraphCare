```python
import pandas as pd

notes = pd.read_csv('/data/physionet.org/files/mimiciii/1.4/NOTEEVENTS.csv')
```

```python
import pickle

with open('/data/pj20/exp_data/ccscm_ccsproc/sample_dataset_drugrec_th015.pkl', 'rb') as f:
    dataset_0 = pickle.load(f)

with open('/data/pj20/exp_data/ccscm_ccsproc_atc3/sample_dataset_mortality_cpd.pkl', 'rb') as f:
    dataset_1 = pickle.load(f)
```

```python
patient_id_set = set()

for patient in dataset_0:
    patient_id_set.add(int(patient['patient_id']))

for patient in dataset_1:
    patient_id_set.add(int(patient['patient_id']))
```

```python
filtered_df = notes[notes['SUBJECT_ID'].isin(patient_id_set)]
```

```python
filtered_df
```

```python
# Assuming the dataframe is named "df"
grouped = filtered_df.groupby('SUBJECT_ID')['TEXT'].apply(lambda x: ' '.join(x)).reset_index()

# Create a dictionary with the combined text for each subject
subject_text_dict = {row['SUBJECT_ID']: row['TEXT'] for _, row in grouped.iterrows()}
```

```python
len(subject_text_dict)
```

```python
import json

with open('./subject_text_dict.json', 'w') as f:
    json.dump(subject_text_dict, f, indent=6)
```
