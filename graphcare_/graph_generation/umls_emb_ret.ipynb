```python
import json

with open('../../graphs/condition/ICD9CM_base/id2ent_new.json', 'r') as f:
    id2ent = json.load(f)

with open('../../graphs/condition/ICD9CM_base/ent2word.json', 'r') as f:
    ent2word = json.load(f)   
```

```python
from get_emb import embedding_retriever
import numpy as np
from tqdm import tqdm
import pickle

## get embedding for procedure entities
ent_emb = []

for i in tqdm(range(len(id2ent.keys()))):
    ent = id2ent[str(i)]
    word = ent2word[ent]
    embedding = embedding_retriever(term=word)
    embedding = np.array(embedding)
    ent_emb.append(embedding)

stacked_embedding = np.vstack(ent_emb)


emb_pkl = f"../../graphs/condition/ICD9CM_base/entity_embedding_new.pkl"

with open(emb_pkl, "wb") as file:
    pickle.dump(stacked_embedding, file)
```

```python
np.save(arr=stacked_embedding, file='../../graphs/condition/ICD9CM_base/kg_ent_emb_new.npy')
```

```python
with open('../../graphs/condition/ICD9CM_base/miss_ent2id.json', 'r') as f:
    miss_ent2id = json.load(f)
```

```python
import pandas as pd

df_1 = pd.read_csv('../../resources/ICD9CM.csv')
df_2 = pd.read_csv('../../resources/ICD9PROC.csv')
code_to_name = {}

# iterate over rows in the dataframe
for index, row in df_1.iterrows():
    # modify the code by removing the period
    code = row['code'].replace('.', '')
    # add the code and name to the dictionary
    code_to_name[code] = row['name']

for index, row in df_2.iterrows():
    # modify the code by removing the period
    code = row['code'].replace('.', '')
    # add the code and name to the dictionary
    code_to_name[code] = row['name']
```

```python
miss_id2ent = {value: key for key, value in miss_ent2id.items()}
```

```python
miss_ent_emb = {}

for i in tqdm(range(len(miss_id2ent))):
    word = code_to_name[miss_id2ent[i]]
    embedding = embedding_retriever(term=word)
    embedding = np.array(embedding)
    ent_emb.append(embedding)

stacked_embedding = np.vstack(ent_emb)

np.save(arr=stacked_embedding, file='../../graphs/condition/ICD9CM_base/kg_ent_emb_miss.npy')
```
